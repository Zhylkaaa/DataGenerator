```python
lastnames = []
with open("list-of-polish-first-and-last-names/polish_surnames.txt") as f:
    lines = f.readlines()
    for name in lines:
        lastnames.append(name[:-1])
firstnames = dict()
firstnames['female'] = []
firstnames['male'] = []

with open("list-of-polish-first-and-last-names/polish_female_firstnames.txt") as f:
    lines = f.readlines()
    for name in lines:
        firstnames['female'].append(name[:-1])

with open("list-of-polish-first-and-last-names/polish_male_firstnames.txt") as f:
    lines = f.readlines()
    for name in lines:
        firstnames['male'].append(name[:-1])
```

```python
persons = []
polish_leters = {
    'ę': 'e',
    'ą': 'a',
    'ń': 'n',
    'ń': 'n',
    'ś': 's',
    'ć': 'c',
    'ł': 'l',
    'ż': 'z',
    'ź': 'z',
    'ó': 'o',
    'ü': 'u',
    'ö': 'o',
    'ę': 'e'
}
for _ in range(8000):
    
    gender = np.random.randint(0, 2)
    is_student = np.random.randint(0, 2)
    gender = 'male' if gender == 1 else 'female'
    person_first_name = firstnames[gender][np.random.randint(0, len(firstnames[gender]))] 
    person_last_name = lastnames[np.random.randint(0, len(lastnames))]
    if(gender == 'female'):
        last_name_split = person_last_name.split('-')
            
        if(last_name_split[0][-3:] == 'ski'):
            last_name_split[0] = last_name_split[0][:-3] + 'ska'
        
        if(len(last_name_split) == 2):
            if(last_name_split[1][-3:] == 'ski'):
                last_name_split[1] = last_name_split[1][:-3] + 'ska'
        
        if(len(last_name_split) == 2):
            person_last_name = '-'.join(last_name_split)
        else:
            person_last_name = last_name_split[0]
        
    person_last_name = person_last_name[0].upper() + person_last_name[1:]
    person_email = person_first_name.lower() + '_' + person_last_name.lower() + '@cokolwiek.com'
    person_email = person_email.replace('-', '_')
    for l in polish_leters:
        person_email = person_email.replace(l, polish_leters[l])
    
    person_phone = '+48' + str(np.random.randint(10**8, 10**9))
    
    person = [person_first_name, person_last_name, person_email, person_phone, is_student]
    persons.append(person)
print(polish_leters)
```

```python
names_df = pd.DataFrame(persons, columns=['FirstName', 'LastName', 'Email', 'Phone', 'IsStudent'])
```

```python
replaces = {
    '.': '',
    '\'': '',
    '&': 'and'
}
company_mails = []
company_phone = []

for name in company_names:
    for r in replaces:
        name = name.replace(r, replaces[r])
    company_mails.append('.'.join(name.split()).lower() + '@company.com')
    company_phone.append('+48' + str(np.random.randint(10**8, 10**9)))
```
